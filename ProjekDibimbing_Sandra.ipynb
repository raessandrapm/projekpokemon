```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
%matplotlib inline
```

```python
df = pd.read_csv('/content/gdrive/MyDrive/Pokemon.csv')
```

```python
df.shape
df.info()
```

```python
df= df.rename(columns={"Name":"nama", "Type 1":"tipe_1","Type 2":"tipe_2", "Total":"total",
                       "HP":"hp", "Attack":"attack","Defence":"defence","Sp. Atk":"sp_atk","Sp. Def":"sp_def",
                       "Speed":"speed","Generation":"generation","Legendary":"legendary",})
```

```python
df.head()
```

```python
df.describe()
```

```python
#menghitung pokemon tiap generasi
sns.countplot(x=df['generation'])
plt.title('Pokemon Tiap Generasi ')
```

```python
#menghitung pokemon legendaris
sns.countplot(x=df[df['legendary']== 1]['generation'])
plt.title('Pokemon Legendaris ')
```

```python
df[df['legendary'] == 1]['generation'].value_counts(sort=False)
```

```python
#pokemon yang paling kuat
df[df['hp'] == df['hp'].max()][['nama', 'hp']]
```

```python
#pokemon paling lemah
df[df['hp'] == df['hp'].min()][['nama', 'hp']]
```

```python
#pokemon yang paling kuat serangannya
df[df['attack'] == df['attack'].max()][['nama', 'attack']]
```

```python
df[df['speed'] == df['speed'].max()][['nama', 'speed']]
```

```python
stats = ['hp', 'attack', 'Defense', 'speed', 'sp_atk', 'sp_def']
for stat in stats:
    print("Stats {0}: ".format(stat))
    display(df[df[stat] == df[stat].max()][['nama', stat]])
```
